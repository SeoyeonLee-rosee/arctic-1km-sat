```python
#import iris
#import iris.quickplot as qplt
#import iris.plot as iplt
#import iris.coord_categorisation
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
import numpy as np
import numpy.ma as ma
import time
import statsmodels.api as sm
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
import matplotlib.ticker as mticker
```

```python
NAME = pd.read_csv("../data/CRU/station/NAME.csv", encoding='cp949')
print(NAME.shape)
NAME.head()
```

```python
All = NAME[NAME['W'] >= 1]
WO = NAME[NAME['W'] == 0]
```

```python
fig = plt.figure(figsize=(10,10))
#ax.scatter(WO['LON'], WO['LAT'],  marker='o', color='k')
#ax.scatter(All['LON'], All['LAT'], marker='x', color='r')

ax =plt.axes(projection=ccrs.NorthPolarStereo())
ax.set_extent([180,-179,55,90], crs=ccrs.PlateCarree())
ax.gridlines(color='lightgray', linestyle='--', linewidth=0.5)
plt.gca().coastlines(color='gray')

ax.scatter(WO['LON'], WO['LAT'], transform=ccrs.PlateCarree(), marker='o', color='k', s=5)
ax.scatter(All['LON'], All['LAT'], transform=ccrs.PlateCarree(), marker='o', color='r', s=10)


#plt.show();
plt.savefig('map.png', dpi=300)
```
